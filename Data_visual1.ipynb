## the use of Matplotlib in python visualization


```python
import matplotlib.pyplot as plt
```

```python
# Example data to use
x = [1, 2, 3, 4, 5]
y = [10, 5, 8, 12, 6]
```

```python
# creating a basic plot
plt.plot(x, y)
plt.show()
```

```python
## let us now add labels to the plot
plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot')
plt.show()
```

```python
## Customizing the plot we have
plt.plot(x, y, color='b', linestyle='--', marker='o', label='Data Points')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Customized Line Plot')
plt.legend()
plt.grid(True)
plt.xlim(0, 6)  # Set the x-axis range from 0 to 6
plt.ylim(0, 15)  # Set the y-axis range from 0 to 15
plt.show()
```

```python
## Using different plottngs
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 5, 8, 12, 6]

plt.bar(x, y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
## Plotting multiple datasets
x = [1, 2, 3, 4, 5]
y1 = [10, 5, 8, 12, 6]
y2 = [8, 6, 7, 9, 5]

plt.plot(x, y1, label='Data 1')
plt.plot(x, y2, label='Data 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Data Sets')
plt.legend()
plt.show()
```
